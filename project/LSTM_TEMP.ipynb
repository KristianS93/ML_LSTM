```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

Convert csv to dataframe

```python
df = pd.read_csv(csv_path)
df
```

[5::6] = with 10 min intervals we take the data point at every hour

```python
df = df[5::6]
df
```

```python
df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
df[:26]
```

26 her bliver de første 26 timer 

```python
temp = df['T (degC)']
temp.plot()
```

The idea of the next function is to split the training set, such that we we for the training set can skip the next time series 
so we need to create our x matrice and y vector
Format:
[[1], [2], [3], [4], [5]] [6]
[[2], [3], [4], [5], [6]] [6]
[[3], [4], [5], [6], [7]] [8]
Based on the last 5 hours we predict the next temperature for the next hour.

```python
def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(temp, WINDOW_SIZE)
X1.shape, y1.shape
print(X1[0])
print(y1[0])
```

Now we have to split the data in test and train data

```python
X_train, y_train = X1[:60000], y1[:60000]
X_val, y_val = X1[60000:65000], y1[60000:65000]
X_test, y_test = X1[65000:], y1[65000:]
print(X_train[0])
X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape
```

Now we import tensor flow
The model will have an input layer, allso called batch that matches the data we are "parsing" into the neural network. 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5,1)))
model1.add(LSTM(64))
model1.add(Dense(8, activation='relu'))
model1.add(Dense(1, activation='linear'))
model1.summary()
```

Save model

```python
cp = ModelCheckpoint('model1/model.keras', save_best_only=True)
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

Now we train the model:

```python
model1.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, callbacks=[cp])
```

Now load the best model:

```python
from tensorflow.keras.models import load_model
model1 = load_model('model1/model.keras')
```

Now we have to compare with the actual test data

```python
train_predictions = model1.predict(X_train).flatten()
train_results = pd.DataFrame(data={'Train Predictions':train_predictions, 'Actuals':y_train})
train_results
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results['Train Predictions'][50:100])
plt.plot(train_results['Actuals'][50:100])
```

```python
val_predictions = model1.predict(X_val).flatten()
val_results = pd.DataFrame(data={'Val Predictions':val_predictions, 'Actuals':y_val})
val_results
```
